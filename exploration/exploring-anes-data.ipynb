## Let's calculate some statistics individually on fields

## Central Tendency

### the mean

we get `<NA>` due to missing values

```r
mean(anes$thermometer_clinton)
```

so we need to instruct R to ignore `<NA>` values

```r
mean(anes$thermometer_clinton, na.rm=T)
```

### the median

```r
median(anes$thermometer_clinton, na.rm=T)
```

### the mode
which is typically applied to ordinal data (strings), nominal (qualitative measures)

```r
library(DescTools)
```

```r
Mode(anes$thermometer_clinton)
```

```r
Mode(anes$party_id)
```

```r
Mode(anes$better_worse_1year)
```

it can be useful to look at `table()` of nominal / ordinal data

```r
table(anes$party_id)
```

```r
table(anes$ideology)
```

## Dispersion

### range
generally, max() - min()

```r
max(anes$thermometer_clinton, na.rm=T)
```

```r
min(anes$thermometer_clinton, na.rm=T)
```

### standard deviation

```r
sd(anes$thermometer_clinton, na.rm=T)
```

### variance

```r
var(anes$thermometer_clinton, na.rm=T)
```

## Scatter Plot

```r
# Scatter plot(x, y)
plot(anes$ideology, 
     anes$thermometer_clinton,
     main="Clinton Thermometer Response By Political Ideology",
     xlab="Political Ideology",
     ylab="Thermometer Score",
     pch=19,
     col = rgb(red = 1, green = 0, blue = 0, alpha = 0.1))
```
